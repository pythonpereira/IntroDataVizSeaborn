```python
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina' # This works in retina Macs only
sns.set()
sns.set_context('talk')
```

# Distribuciones univariadas

```python
tips = sns.load_dataset('tips')
tips[:10]
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, rug=True)
```

# Distribuciones bivariadas

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

# Scatterplot categóricos

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', data=tips)
```

# Estimación estadística en las categorías

```python
titanic = sns.load_dataset('titanic')
titanic[:10]
```

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.countplot(x='embark_town', data=titanic)
```

```python
sns.pointplot(x='class', y='survived', hue='sex', data=titanic)
```

# Relaciones lineales

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='size', y='tip', data=tips)
```

```python
sns.lmplot(x='size', y='tip', data=tips, x_jitter=0.1)
```

```python
sns.lmplot(x='size', y='tip', data=tips, x_estimator=np.mean)
```

## Cuarteto de Anscombe

```python
anscombe = sns.load_dataset('anscombe')
anscombe[:10]
```

```python
anscombe_1 = anscombe[anscombe.dataset == 'I']
sns.lmplot(x='x', y='y', data=anscombe_1, ci=None)
```

```python
anscombe_2 = anscombe[anscombe.dataset == 'II']
sns.lmplot(x='x', y='y', data=anscombe_2, ci=None)
```

```python
sns.lmplot(x='x', y='y', data=anscombe_2, ci=None, order=2)
```

```python
anscombe_3 = anscombe[anscombe.dataset == 'III']
sns.lmplot(x='x', y='y', data=anscombe_3, ci=None)
```

```python
# Requires statsmodels
sns.lmplot(x='x', y='y', data=anscombe_3, ci=None, robust=True)
```

# ¿Cómo la relación entre dos variables cambia respecto a otras?

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='time', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='time', row='sex', data=tips)
```
